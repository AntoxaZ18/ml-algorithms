```python
import pandas as pd
```

```python
TRUNC_CSV = 'trunc.csv'

#load dataframe from link and skip first row
df = pd.read_csv('https://www.geos.ed.ac.uk/~weather/jcmb_ws/JCMB_2018_Mar.csv', skiprows = 1)

```

```python
df.head()
```

```python
units = df.iloc[1].fillna('')   #save measurement units and metricks to separate dict 
parameters = df.iloc[0].fillna('')

column_units = {column_name: (meas_untis, metrick) for (column_name, meas_untis, metrick) in zip(df.columns, units, parameters)}


#drop unused rows and reset index
df = df.drop([0,1,2])
df = df.reset_index(drop=True)
```

```python
#save first 1000 samples to csv
df[:1000].to_csv(TRUNC_CSV)
```
